## (3)
I assume a quite high starting value of the habit to capture the severity of the great depression.

```julia
function HABITScalc(df, δ, θ, γ)
    h = zeros(length(df[:C_ND]))
    s = zeros(length(df[:C_ND]))
    
    h[1] = df[:C_ND][1]*3/4
    
    for i in 2:length(h)
        h[i] = (1-δ)*h[i-1] + δ*df[:C_ND][i-1]
        s[i] = df[:C_ND][i] - θ*h[i]
    end
    s = s[2:end]
    M = (exp(diff(log(s)))).^(-γ)
    return [mean(M), std(M)]
end
```

```julia
result = [HABITScalc(df, δ, θ, γ) for δ in [0.1, 0.05, 0.01], θ in [0.9, 0.95, 0.99], γ in [2, 4, 6, 8, 10, 15, 20]]
scatter!([result[:][i][1] for i in 1:length(result)], [result[:][i][2] for i in 1:length(result)], label="Habits")
```

The lowest $\gamma=4$ I could find was with $\delta=0.01$ and $\theta=0.99$:

```julia
scatter!([HABITScalc(df, 0.01, 0.99, 4)[1]],[HABITScalc(df, 0.01, 0.99, 4)[2]], label="\\gamma=4, \\delta=0.01, \\theta=0.99")
```

Not too bad actually!

# Question 3

```julia
using Optim
```

```julia
function g(c,r,γ) 
    return 0.998 * (c).^(-γ) .* r - 1
end

#operates on data set!
function gn(c,r,γ)
    i = 1
    result = g(c[i],r[i,:],γ)
    
    for i in 2:size(c)[1]
        result .+= g(c[i],r[i,:],γ)
    end
    return result/size(c)[1]
end

#Wrapper creates a closure around the data set provided
function Qn_wrapper(c,r)
    return γ -> 1/2*(gn(c,r,γ)'*gn(c,r,γ))[1,1]
end

function Qn_wrapper(c,r,What)
    return γ -> 1/2*((gn(c,r,γ))'*What*gn(c,r,γ))[1,1]
end

#Calculates What
function What(c,r,γ)
    result = zeros(Float64, size(r)[2], size(r)[2])
    for i in 1:length(c)
        result .+= g(c[i],r[i,:],γ)*g(c[i],r[i,:],γ)'
    end
    return inv(result/length(c)[1])
end

#creates a closure around the data set
function gn_wrapper(c,r)
    return γ -> gn(c,r,γ)
end
```

```julia
table = DataFrame(Name = String[], γfirst = Real[], γsecond = Real[], J = Real[], pvalue = Real[]);
```

## (a)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[:Rm]./100+1)
Qn = Qn_wrapper(c,r)
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
push!(table, ["excess market" optres.minimizer NA NA NA])
```

No need for a second stage or a J-statistic in the just-identified setup.

## (d)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[1:end-1, [:rRf, :rRm]])./100+1
Qn = Qn_wrapper(c,r)
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
γfirst = optres.minimizer[1]
Whatfirst = What(c,r,γfirst)
QnWhatfirst = Qn_wrapper(c,r,Whatfirst)
optressecond = optimize(QnWhatfirst, [γfirst], BFGS(), Optim.Options(g_tol=1e-12, autodiff=true))
γsecond = optressecond.minimizer[1]
Jn = 2*length(c)*QnWhatfirst(γsecond)
p = 1-cdf(Chisq(size(r)[2]-1),Jn)
push!(table, ["real market and riskfree" γfirst γsecond Jn p])
```
