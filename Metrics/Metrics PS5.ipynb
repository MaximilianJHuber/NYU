```julia
using Distributions
using Plots
pyplot()
```

```julia
function sim(;n=200, β=1, σeu=0.5, R=0.5)
    γ = sqrt(R/(1-R))
    e = rand(MvNormal([0., 0],[1 0.5; 0.5 1]), n)'
    Z = rand(Normal(),n)
    X = Z*γ + e[:,2]
    y = X*β + e[:,1]
        
    βols = (X'X)\(X'y)
    β2sls = 1/(X'Z.*1/(Z'Z)[1,1].*Z'X)[1,1].*X'Z.*1/(Z'Z)[1,1].*Z'y
    
    return vcat(βols, β2sls)
end
```

```julia
simulations = SharedArray(Float64, 10000, 2)
for i in 1:10000
    simulations[i,:] = sim()
end;
```

d.
--
a. and b. show that the 2SLS, which only uses the "clean" variation in X (induced by Z) gives an unbiased estimate of $\beta$, whereas the OLS consistently overestimates.
But in d., where Z becomes a very weak instrument for X, the estimate is highly dispersed. Yet it's median is still close to the true value.
