## (3)
I assume a quite high starting value of the habit to capture the severity of the great depression.

```julia
function HABITScalc(df, δ, θ, γ)
    h = zeros(length(df[:C_ND]))
    s = zeros(length(df[:C_ND]))
    
    h[1] = df[:C_ND][1]*3/4
    
    for i in 2:length(h)
        h[i] = (1-δ)*h[i-1] + δ*df[:C_ND][i-1]
        s[i] = df[:C_ND][i] - θ*h[i]
    end
    s = s[2:end]
    M = (exp(diff(log(s)))).^(-γ)
    return [mean(M), std(M)]
end
```

```julia
result = [HABITScalc(df, δ, θ, γ) for δ in [0.1, 0.05, 0.01], θ in [0.9, 0.95, 0.99], γ in [2, 4, 6, 8, 10, 15, 20]]
scatter!([result[:][i][1] for i in 1:length(result)], [result[:][i][2] for i in 1:length(result)], label="Habits")
```

The lowest $\gamma=4$ I could find was with $\delta=0.01$ and $\theta=0.99$:

```julia
scatter!([HABITScalc(df, 0.01, 0.99, 4)[1]],[HABITScalc(df, 0.01, 0.99, 4)[2]], label="\\gamma=4, \\delta=0.01, \\theta=0.99")
```

Not too bad actually!

# Question 3

```julia
using Optim
```

```julia
function g(c,r,γ, RHS) 
    return 0.998 * (c).^(-γ) .* r - RHS
end

#operates on data set!
function gn(c,r,γ, RHS)
    i = 1
    result = g(c[i],r[i,:],γ, RHS)
    
    for i in 2:size(c)[1]
        result .+= g(c[i],r[i,:],γ, RHS)
    end
    return result/size(c)[1]
end

#Wrapper creates a closure around the data set provided
function Qn_wrapper(c,r,RHS)
    return γ -> 1/2*(gn(c,r,γ,RHS)'*gn(c,r,γ,RHS))[1,1]
end

function Qn_wrapper(c,r,What,RHS)
    return γ -> 1/2*((gn(c,r,γ,RHS))'*What*gn(c,r,γ,RHS))[1,1]
end

#Calculates What
function What(c,r,γ,RHS)
    result = zeros(Float64, size(r)[2], size(r)[2])
    for i in 1:length(c)
        result .+= g(c[i],r[i,:],γ,RHS)*g(c[i],r[i,:],γ,RHS)'
    end
    return inv(result/length(c)[1])
end
```

```julia
table = DataFrame(Name = String[], γfirst = Real[], γsecond = Real[], J = Real[], pvalue = Real[]);
```

## (a)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[:Rm]./100+1)
Qn = Qn_wrapper(c,r,0) #mind: excess return => RHS = 0
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
push!(table, ["excess market" optres.minimizer NA NA NA])
```

No need for a second stage or a J-statistic in the just-identified setup.
## (b)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[:rRm][1:end-1]./100+1)
Qn = Qn_wrapper(c,r,0)
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
push!(table, ["real market" optres.minimizer NA NA NA])
```

## (c)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[:rRf][1:end-1]./100+1)
Qn = Qn_wrapper(c,r,1)
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
push!(table, ["real riskfree" optres.minimizer NA NA NA])
```

## (d)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[1:end-1, [:rRf, :rRm]])./100+1
Qn = Qn_wrapper(c,r,[1, 0])
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
γfirst = optres.minimizer[1]
Whatfirst = What(c,r,γfirst,[1, 1])
QnWhatfirst = Qn_wrapper(c,r,Whatfirst,[1, 1])
optressecond = optimize(QnWhatfirst, [γfirst], BFGS(), Optim.Options(g_tol=1e-12, autodiff=true))
γsecond = optressecond.minimizer[1]
Jn = 2*length(c)*QnWhatfirst(γsecond)
p = 1-cdf(Chisq(size(r)[2]-1),Jn)
push!(table, ["real market and riskfree" γfirst γsecond Jn p])
```

## (e)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[1:end-1, [:rRf, :rExr]])./100+1
Qn = Qn_wrapper(c,r,[1, 0])
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
γfirst = optres.minimizer[1]
Whatfirst = What(c,r,γfirst,[1, 0])
QnWhatfirst = Qn_wrapper(c,r,Whatfirst,[1, 0])
optressecond = optimize(QnWhatfirst, [γfirst], BFGS(), Optim.Options(g_tol=1e-12, autodiff=true))
γsecond = optressecond.minimizer[1]
Jn = 2*length(c)*QnWhatfirst(γsecond)
p = 1-cdf(Chisq(size(r)[2]-1),Jn)
push!(table, ["excess market and real riskfree" γfirst γsecond Jn p])
```

## (f)

```julia
c = Array(df[:gC_ND][1:end-1]./100+1)
r = Array(df[1:end-1, [:rRf, :rRm, :rs1, :rs2, :rs3, :rb1, :rb2, :rb3]])./100+1
Qn = Qn_wrapper(c,r,[1, 1, 0, 0, 0, 0, 0, 0])
optres = optimize(Qn, [10.], BFGS(), Optim.Options(g_tol=1e-14, autodiff=true))
γfirst = optres.minimizer[1]
Whatfirst = What(c,r,γfirst,[1, 1, 0, 0, 0, 0, 0, 0])
QnWhatfirst = Qn_wrapper(c,r,Whatfirst,[1, 1, 0, 0, 0, 0, 0, 0])
optressecond = optimize(QnWhatfirst, [γfirst], BFGS(), Optim.Options(g_tol=1e-12, autodiff=true))
γsecond = optressecond.minimizer[1]
Jn = 2*length(c)*QnWhatfirst(γsecond)
p = 1-cdf(Chisq(size(r)[2]-1),Jn)
push!(table, ["8 assets" γfirst γsecond Jn p])
```

```julia
h = zeros(length(df[:C_ND]))
s = zeros(length(df[:C_ND]))

s[1] = log(Sbar*5/3)

for i in 2:length(h)-1
    s[i] = (1 - ϕ) * log(Sbar) + ϕ * s[i-1] + λ(s[i-1]) * (log(1+df[i, :gC_ND]/100) - gmean/100)
end
```
